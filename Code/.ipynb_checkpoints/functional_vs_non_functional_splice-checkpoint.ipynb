```python
import pandas as pd
import gzip
from tqdm import tqdm

# Define the file path
file_path = '/odinn/data/extdata/GTEx/v8/GTEx_Analysis_2017-06-05_v8_STARv2.5.3a_junctions.gct.gz'  # Replace 'your_file.txt.gz' with the actual file path

# Create an empty DataFrame to store the filtered data
columns = ['Name','Description']  # Replace with your actual column names
df = pd.DataFrame(columns=columns)

# Open the Gzip-compressed file and process it line by line

with gzip.open(file_path, 'rt') as file:
    for i,line in tqdm(enumerate(file)):
        if i>2:
            # Split the line into columns using tab as the separator
            columns = line.strip().split('\t')

            # Check if the row has a single read in all columns combined
            #if sum(map(int, columns[2:])) == 1:
            tmp = np.array(columns[2:]).astype(int)
            if np.any(tmp==1) and np.all(tmp<=1):
                # Add the row to the DataFrame
                df = df.append(pd.Series(columns[:2], index=['Name','Description']), ignore_index=True)

# Display the resulting DataFrame
print(df)
```

```python
import pyfastx
import gffutils
```

```python
gtf_gencode = gffutils.FeatureDB('/odinn/tmp/benediktj/Data/SplicePrediction-GTEX-V8/gencode.v42.annotation.db')
fasta = pyfastx.Fasta('/odinn/tmp/benediktj/Data/SplicePrediction-GTEX-V8/GRCh38.p13.genome.fa')
```

```python
chromosomes = [f'chr{i}' for i in range(1,23)]+['chrX']
```

```python
strands = ['+','-']
```

```python
lengths = {}
for chrom in chromosomes:
    lengths[chrom] = len(fasta[chrom])
```

```python
import numpy as np
```

```python
lengths
```

```python
n_samples = 10000
posistions = []
last = 0
while len(posistions)<n_samples:
    chrom = np.random.choice(chromosomes)
    strand = np.random.choice(strands)
    length = lengths[chrom]
    pos = np.random.randint(1, high=length-2)
    if strand=='+':
        value = fasta[str(chrom)][pos:pos+2].seq
    else:
        value = fasta[str(chrom)][pos:pos+2].antisense
    if value == 'AG' or value == 'GT':
        posistions.append([chrom,str(pos),strand])
    if len(posistions)>last:
        last = len(posistions)
        if last%25==0:
            print(last)
```
